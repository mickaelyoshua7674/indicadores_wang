## NORMALIZANDO DADOS

```python
#UTILIZANDO NORMALIZAÇÃO
```

```python
tab_wang['Posição de Caixa'] = (tab_wang['Posição de Caixa'] - round(tab_wang['Liquidez Imediata'].mean(),2)) / round(np.std(tab_wang['Posição de Caixa']),2)

tab_wang['Liquidez Imediata'] = (tab_wang['Liquidez Imediata'] - round(tab_wang['Posição Operacional'].mean(),2)) / round(np.std(tab_wang['Liquidez Imediata']),2)

tab_wang['Liquidez Corrente'] = (tab_wang['Liquidez Corrente'] - round(tab_wang['Superávit per capita'].mean(),2)) / round(np.std(tab_wang['Liquidez Corrente']),2)

tab_wang['Posição Operacional'] = (tab_wang['Posição Operacional'] - round(tab_wang['Índice de Ativos Líquidos'].mean(),2)) / round(np.std(tab_wang['Posição Operacional']),2)

tab_wang['Superávit per capita'] = (tab_wang['Superávit per capita'] - round(tab_wang['Obrigações de LP'].mean(),2)) / round(np.std(tab_wang['Superávit per capita']),2)

tab_wang['índice de Ativos Líquidos'] = (tab_wang['Índice de Ativos Líquidos'] - round(tab_wang['Obrigações de LP per capita'].mean(),2)) / round(np.std(tab_wang['Índice de Ativos Líquidos']),2)

tab_wang['Obrigações de LP'] = (tab_wang['Obrigações de LP'] - round(tab_wang['Tributos per capita'].mean(),2)) / round(np.std(tab_wang['Obrigações de LP']),2)

tab_wang['Obrigações de LP per capita'] = (tab_wang['Obrigações de LP per capita'] - round(tab_wang['Receita Per capita'].mean(),2)) / round(np.std(tab_wang['Obrigações de LP per capita']),2)

tab_wang['Tributos per capita'] = (tab_wang['Tributos per capita'] - round(tab_wang['Despesa per capita'].mean(),2)) / round(np.std(tab_wang['Tributos per capita']),2)

tab_wang['Receita Per capita'] = (tab_wang['Receita Per capita'])/ round(np.std(tab_wang['Receita Per capita']),2)

tab_wang['Despesa per capita'] = (tab_wang['Despesa per capita'])/ round(np.std(tab_wang['Despesa per capita']),2)
```

## DEIXANDO DUAS CASAS DECIMAIS

```python
tab_wang['Solvência de Caixa'] = (tab_wang['Posição de Caixa'] + tab_wang['Liquidez Corrente'] +  tab_wang['Liquidez Imediata'])/3

tab_wang['Solvência Orçamentária'] = (tab_wang['Posição Operacional'] + tab_wang['Superávit per capita'])/2

tab_wang['Solvência de Longo Prazo'] = -(tab_wang['Obrigações de LP'] + tab_wang['Obrigações de LP per capita'] +  tab_wang['Índice de Ativos Líquidos'])/3

tab_wang['Solvência de Serviços'] = (tab_wang['Tributos per capita'] + tab_wang['Despesa per capita'] +  tab_wang['Receita Per capita'])/3

##########################################

tab_wang['CFG (Wang) Geral'] = (tab_wang['Solvência de Caixa'] + tab_wang['Solvência de Serviços'] + tab_wang['Solvência Orçamentária'] +  tab_wang['Solvência de Longo Prazo'])/4
```
